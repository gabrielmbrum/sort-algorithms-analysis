```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dados fornecidos
n = [10, 100, 1000, 10000, 100000, 1000000]  # Tamanhos do array
sorted_times = [0.000001, 0.000001, 0.000001, 0.000011, 0.000110, 0.000972]
unsorted_times = [0.0000001, 0.000012, 0.001128, 0.117072, 10.622649, 1091.755737]
randomized_times = [0.0000001, 0.000001, 0.000002, 0.000011, 0.000101, 0.000986]


```
